what is the difference between image+100 and cv.add(image,100) ?

cv.add(image,100) = min((image+100),255) 








```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

Original_image  = cv.imread('../images/pahansith.png', cv.IMREAD_GRAYSCALE)

def Brightner(image, shift):
    rows = image.shape[0]
    columns = image.shape[1]
    output = np.zeros(image.shape,image.dtype)
    for i in range(rows): 
        for j in range(columns):
             if (image[i][j]+shift > 255):  output[i][j]=255 
             else: output[i,j]=image[i,j]+shift 
    return output

output = Brightner(Original_image,100)

fig1 , fig = plt.subplots(1,2)
fig[0].imshow(Original_image,cmap = 'gray')
fig[0].set_title("Original")
fig[1].imshow(output, cmap = 'gray')
fig[1].set_title("Brightned")

"""
cv.namedWindow("Image" , cv.WINDOW_FULLSCREEN)
cv.imshow("Image", image)
cv.waitKey(0)
cv.destroyAllWindows() """
```

```python
import cv2 as cv
import matplotlib.pyplot as plt

Original_image = cv.imread('../images/pahansith.png',cv.IMREAD_COLOR)
Converted_image = cv.cvtColor(Original_image, cv.COLOR_BGR2RGB)
#brightned_image = cv.add(Converted_image,100)

def Brightner(image, shift):
    rows = image.shape[0]
    columns = image.shape[1]
    planes = image.shape[2]
    print(planes)
    output = np.zeros(image.shape,image.dtype)
    for i in range(rows): 
        for j in range(columns):
            for k in range(planes):
                if (image[i][j][k]+shift > 255):  output[i][j][k]=255 
                else: output[i,j,k]=image[i,j,k]+shift 
    return output
Brightned_image = Brightner(Converted_image,100)
print(Converted_image)
print(Brightned_image)
fig1 , fig = plt.subplots(1,3)
fig[0].imshow(Original_image)
fig[0].set_title("Original")
fig[1].imshow(Converted_image)
fig[1].set_title("Converted")
fig[2].imshow(Brightned_image)
fig[2].set_title("Brightned")
```
